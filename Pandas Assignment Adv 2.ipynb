```python
# Q6. What do you understand about the windows function in pandas and list the types of windows
# functions?

"""
In Pandas, the window functions provide a way to perform rolling computations or calculations over
a specified window or range of values in a Series or DataFrame. These functions allow you to apply
operations on a sliding window of data, typically used for calculations such as moving averages,
cumulative sums, or other window-based statistics.

The window functions in Pandas can be categorized into three main types based on the type of calculations
they perform:

(*) Aggregating window functions: These functions aggregate values within a defined window to produce a 
                                single value. Examples include rolling.mean(), rolling.sum(), rolling.min(),
                                rolling.max(), rolling.median(), rolling.std(), etc.

(*) Expanding window functions: These functions accumulate or expand calculations over time as the window grows.
                                They compute statistics for all preceding elements up to the current element. 
                                Examples include expanding.mean(), expanding.sum(), expanding.min(), 
                                expanding.max(), etc.

(*)Rolling window functions: These functions calculate statistics on a rolling window of fixed size or time period.
                                The window slides across the data, recalculating the statistics for each 
                                overlapping window. Examples include rolling.apply(), rolling.aggregate(), 
                                rolling.transform(), rolling.corr(), rolling.cov(), etc.
```

```python
#Q8. Write a Python program that takes in two dates as input (in the format YYYY-MM-DD) and
#calculates the difference between them in days, hours, and minutes using Pandas time delta. The
#program should prompt the user to enter the dates and display the result.


import pandas as pd

dt1 = input("Enter the first date (YYYY-MM-DD): ")   #asking user for Date1
dt2 = input("Enter the second date (YYYY-MM-DD): ")  #asking user for Date2


timestamp1 = pd.Timestamp(dt1)    # Convert  strings to pandas Timestamp objects
timestamp2 = pd.Timestamp(dt2)


time_diff = timestamp2 - timestamp1

# Extract the days, hours, and minutes from the time difference
days = time_diff.days
hours = time_diff.seconds // 3600
minutes = (time_diff.seconds // 60) % 60

# Display the result
print("Difference between", dt1, "and", dt2, ":")
print("Days:", days)
print("Hours:", hours)
print("Minutes:", minutes)
```

```python
#Q9. Write a Python program that reads a CSV file containing categorical data and converts a specified
#column to a categorical data type. The program should prompt the user to enter the file path, column
#name, and category order, and then display the sorted data.
```

```python
import pandas as pd

file_path = input("Enter the file path: ")    # we will read IPL fantasy file here


df = pd.read_excel(file_path)

column_name = input("Enter  column name: ")
category_order = input("Enter  category order (comma-separated values): ").split(',')


df[column_name] = pd.Categorical(df[column_name], categories=category_order, ordered=True)

# Sort the DataFrame based on the specified column
sorted_df = df.sort_values(by=column_name)

# Display the sorted data
print(sorted_df)
```
